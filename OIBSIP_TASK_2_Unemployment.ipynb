```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
plt.rcParams['figure.figsize']=(15,10)
plt.rcParams['font.size']=15
```

```python
raw=pd.read_csv('Unemployment in India.csv')
raw
```

```python
raw.dropna(inplace=True)
```

```python
raw.describe()
```

```python
raw
```

```python
raw.Region.value_counts()
```

```python
sns.barplot(x=raw.Region.value_counts().index,y=raw.Region.value_counts())
plt.xticks(rotation=80)
plt.show(
)
```

```python
plt.pie(raw.Area.value_counts(),labels=raw.Area.value_counts().index,autopct='%1.1f%%')
plt.title('AREA')
plt.show()
```

```python
#FETCH GUJARAT_DATA

guj=raw[['Region','Employed','Area','unemploy','Participate']]
```

```python
guj_df=guj.Region=='Gujarat'
```

```python
st=guj[guj_df]
st_df=st.sort_values(by=['Employed'])
st_df
```

```python
sns.distplot(st_df['unemploy'])
plt.show()
```

```python
sns.pairplot(st_df,hue='Area')
plt.show()
```

```python
sns.distplot(st_df['Participate'])
```
